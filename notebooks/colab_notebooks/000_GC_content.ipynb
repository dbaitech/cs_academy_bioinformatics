## 3. Reading in the FASTA files
The following function takes in the file path, opens the file and passes it into the `SeqIO.read()` function which reads in and organizes the FASTA info into:
- description (the info about the organism the DNA belongs to)
- seq (the DNA sequence)
- length of the DNA sequence

```python
def get_fasta_info(file_path):
    # Open the FASTA file
    with open(file_path, "r", encoding="utf-8") as fasta_file:
        record = SeqIO.read(fasta_file, "fasta")  # read in the FASTA records
        print(f"Description: {record.description}")
        print(f"Sequence: {record.seq[:20]}...")  # Print first 20 bases
        print(f"Length: {len(record.seq)}\n")
```

You can call `get_fasta_info` and pass in any fasta file path to display a summary of its info:
