## 1. Working with Python Dictionaries
Before we create our function for collecting k-mer counts within a sequence, let's first explore a Python dictionary. Specifically, we will use a `defaultdict`, which is a type of dictionary that automatically assigns a default value to new keys. This is useful for counting because we can increase a count immediately without first checking whether the key already exists.

Below, we have the empty dictionary `test_dictionary`, we'll look at how we can add entries, edit them and access them.

```python
from collections import defaultdict

test_dictionary = defaultdict(int)
print(f"Original empty dictionary: {dict(test_dictionary)}")

# If we see "AT" in the sequence, we can increase its count by 1.
# defaultdict automatically creates "AT" with a value of 0 if it
# does not already exist, so we don't need to check first.
test_dictionary["AT"] += 1
print(f"Dictionary after update: {dict(test_dictionary)}")

# Say we saw "AT" again, we can increment it once again:
test_dictionary["AT"] += 1
print(f"Dictionary after update: {dict(test_dictionary)}")

# If we want to access the count of "AT", we simply do:
print(f"Count of 'AT': {test_dictionary["AT"]}")
```
